4.3. Arreglar columna fecha

```python
df_banco['date'].head()
```

Tengo que transformar las palabras en español de los meses a numeros. Primero creo un diccionario con los meses

```python
meses = {
    "enero": "01",
    "febrero": "02",
    "marzo": "03",
    "abril": "04",
    "mayo": "05",
    "junio": "06",
    "julio": "07",
    "agosto": "08",
    "septiembre": "09",
    "octubre": "10",
    "noviembre": "11",
    "diciembre": "12"
}
```

Defino la función de cambio de meses, la aplico y cambio a formato fecha

```python
def cambiar_mes(fecha):
    if pd.notnull(fecha):
        fecha = str(fecha)
        for mes, num in meses.items():
            if mes in fecha:
                fecha = fecha.replace(mes, num)
        return fecha
    return fecha

df_banco['date'] = df_banco['date'].apply(cambiar_mes)

df_banco['date'] = pd.to_datetime(df_banco['date'], format='%d-%m-%Y')
```
